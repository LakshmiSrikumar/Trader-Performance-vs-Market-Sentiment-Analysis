```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

# Part A — Data preparation

```python
# Load datasets

sentiment=pd.read_csv('/kaggle/input/datasets/playokdata/market-sentiment-and-trader-behaviour/fear_greed_index.csv')
trade=pd.read_csv('/kaggle/input/datasets/playokdata/market-sentiment-and-trader-behaviour/historical_data.csv')
```

```python
# Function to summarize dataset
def summarize(df, name):
    print(f"\n{name} SHAPE:", df.shape)
    print(f"\n{name} INFO:")
    print(df.info())
    
    print(f"\n{name} MISSING VALUES:")
    print(df.isna().sum())
    
    print(f"\n{name} DUPLICATES:", df.duplicated().sum())
    
    print(f"\n{name} HEAD:")
    print(df.head())

# Summaries
summarize(sentiment, "SENTIMENT")
summarize(trade, "TRADES")
```

## Sentiment Analysis

### Clean & Prepare Sentiment

```python
# Convert date
sentiment["date"] = pd.to_datetime(sentiment["date"])

# Sort
sentiment = sentiment.sort_values("date")

# Remove duplicates
sentiment = sentiment.drop_duplicates(subset=["date"])

# Encode sentiment
sentiment["sentiment_binary"] = sentiment["classification"].map({
    "Fear": 0,
    "Greed": 1
})

# Check missing dates
full_range = pd.date_range(sentiment["date"].min(), sentiment["date"].max())
missing_dates = full_range.difference(sentiment["date"])

print("Missing dates:", len(missing_dates))
```

### Sentiment EDA

```python
# Distribution
sns.countplot(data=sentiment, x="classification")
plt.title("Fear vs Greed Distribution")
plt.show()

print(sentiment["classification"].value_counts(normalize=True))

# Time series
plt.plot(sentiment["date"], sentiment["sentiment_binary"])
plt.title("Sentiment Over Time (0=Fear, 1=Greed)")
plt.show()
```

### Sentiment Streaks

```python
sentiment["streak_id"] = (
    sentiment["classification"] != sentiment["classification"].shift()
).cumsum()

streaks = sentiment.groupby(
    ["streak_id", "classification"]
).size().reset_index(name="length")

sns.histplot(data=streaks, x="length", hue="classification", bins=20)
plt.title("Sentiment Streak Lengths")
plt.show()
```

## Trader Analysis

### Clean & Prepare Trades

```python
# Copy dataframe
trades = trade.copy()

# Convert timestamp
trades["time"] = pd.to_datetime(
    trades["Timestamp IST"],
    format="%d-%m-%Y %H:%M"
)

# Rename columns
trades = trades.rename(columns={
    "Execution Price": "price",
    "Size Tokens": "size_tokens",
    "Size USD": "size_usd",
    "Closed PnL": "closed_pnl",
    "Account": "account",
    "Side": "side",
    "Direction": "direction"
})

# Numeric conversion
numeric_cols = [
    "price",
    "size_tokens",
    "size_usd",
    "closed_pnl"
]

for col in numeric_cols:
    trades[col] = pd.to_numeric(trades[col], errors="coerce")

# Drop bad rows
trades = trades.dropna(subset=[
    "price",
    "size_tokens",
    "time"
])

# Remove duplicates
trades = trades.drop_duplicates()

print("Cleaned trades shape:", trades.shape)
```

### Feature Engineering

```python
# Long/Short mapping
trades["trade_type"] = trades["side"].map({
    "BUY": "Long",
    "SELL": "Short"
})

# Notional value
trades["notional"] = trades["price"] * trades["size_tokens"]

# Return percentage
trades["return_pct"] = (
    trades["closed_pnl"] / trades["notional"]
)

# Profit flag
trades["is_profit"] = trades["closed_pnl"] > 0

# Extract date
trades["date"] = trades["time"].dt.date
```

### Trader-Level Metrics (Per Account, Per Day)

```python
trader_daily = trades.groupby(
    ["account", "date"]
).agg({
    "closed_pnl": "sum",
    "is_profit": "mean",
    "size_tokens": "count",
    "notional": "mean"
}).reset_index()

trader_daily.columns = [
    "account",
    "date",
    "daily_pnl",
    "win_rate",
    "trade_count",
    "avg_trade_size"
]

print(trader_daily.head())
```

### Market-Level Daily Metrics

```python
daily_metrics = trades.groupby("date").agg({
    "closed_pnl": ["sum", "mean"],
    "account": "nunique",
    "size_tokens": "count",
    "is_profit": "mean"
})

daily_metrics.columns = [
    "total_pnl",
    "avg_pnl",
    "unique_traders",
    "trade_count",
    "win_rate"
]

daily_metrics = daily_metrics.reset_index()
print(daily_metrics)
```

### Additional Behavioral Metrics

```python
# Long / Short ratio
long_short = trades.groupby(
    ["date", "trade_type"]
).size().unstack(fill_value=0)

long_short["long_short_ratio"] = (
    long_short.get("Long", 0) /
    long_short.get("Short", 1)
)

long_short = long_short.reset_index()

# Merge
daily_metrics = daily_metrics.merge(
    long_short,
    on="date",
    how="left"
)
```

# Merge Datasets

```python
# Ensure sentiment date format
sentiment["date"] = pd.to_datetime(
    sentiment["date"]
).dt.date

# Binary encoding
sentiment["sentiment_binary"] = sentiment[
    "classification"
].map({
    "Extreme Fear": 0,
    "Fear": 0,
    "Neutral": 0.5,
    "Greed": 1,
    "Extreme Greed": 1
})

# Merge datasets
merged = daily_metrics.merge(
    sentiment[[
        "date",
        "classification",
        "sentiment_binary",
        "value"
    ]],
    on="date",
    how="inner"
)

print("Merged shape:", merged.shape)
print(merged.head())
```

# Part B — Analysis 

### Preparing Sentiment groups

```python
# Simplify sentiment classes
merged["sentiment_group"] = merged["classification"].replace({
    "Extreme Fear": "Fear",
    "Fear": "Fear",
    "Neutral": "Neutral",
    "Greed": "Greed",
    "Extreme Greed": "Greed"
})

# Optional: remove neutral for direct comparison
analysis_df = merged[
    merged["sentiment_group"] != "Neutral"
].copy()

# analysis_df["sentiment_group"].value_counts()
merged["sentiment_group"].value_counts()
```

## Q1: Does trader performance differ between Fear vs Greed days?

```python
performance_summary = analysis_df.groupby(
    "sentiment_group"
).agg({
    "total_pnl": ["mean", "median"],
    "avg_pnl": ["mean"],
    "win_rate": ["mean"],
    "trade_count": ["mean"]
})

print(performance_summary)
```

```python
sns.boxplot(
    data=analysis_df,
    x="sentiment_group",
    y="total_pnl"
)

plt.title("Total Daily PnL: Fear vs Greed")
plt.show()
```

```python
sns.barplot(
    data=analysis_df,
    x="sentiment_group",
    y="win_rate"
)

plt.title("Win Rate by Sentiment")
plt.ylabel("Average Win Rate")
plt.show()
```

```python
analysis_df["loss_day"] = analysis_df["total_pnl"] < 0

drawdown_proxy = analysis_df.groupby(
    "sentiment_group"
)["loss_day"].mean()

print(drawdown_proxy)
```

## Q2: Do traders behave differently during Fear vs Greed?

```python
sns.barplot(
    data=analysis_df,
    x="sentiment_group",
    y="trade_count"
)

plt.title("Trade Frequency by Sentiment")
plt.show()
```

```python
sns.barplot(
    data=analysis_df,
    x="sentiment_group",
    y="long_short_ratio"
)

plt.title("Long/Short Ratio by Sentiment")
plt.axhline(1, color="red", linestyle="--")
plt.show()
```

```python
daily_position = trades.groupby("date")["notional"].mean().reset_index()
daily_position.columns = ["date", "avg_notional"]

analysis_df = analysis_df.merge(
    daily_position,
    on="date",
    how="left"
)

sns.boxplot(
    data=analysis_df,
    x="sentiment_group",
    y="avg_notional"
)

plt.title("Position Size by Sentiment")
plt.show()
```

### Trader Segmentation

#### Segment 1 — High vs Low Leverage Traders

```python
# Average position size per trader
position_segment = trades.groupby("account")["size_usd"].mean()

# Median threshold
threshold = position_segment.median()

position_segment = pd.DataFrame({
    "avg_position_size": position_segment
})

# Segment
position_segment["position_group"] = np.where(
    position_segment["avg_position_size"] >= threshold,
    "Large Position Traders",
    "Small Position Traders"
)

print(position_segment.head())
```

```python
trades = trades.merge(
    position_segment["position_group"],
    left_on="account",
    right_index=True
)

position_perf = trades.groupby("position_group").agg({
    "closed_pnl": ["mean", "median"],
    "return_pct": "mean",
    "is_profit": "mean",
    "size_usd": "mean"
})

print(position_perf)
```

```python
sns.boxplot(
    data=trades,
    x="position_group",
    y="closed_pnl"
)

plt.title("PnL Distribution by Position Size Segment")
plt.show()
```

#### Insights

Large position traders generate significantly higher average profits per trade, but they also exhibit lower win rates,larger PnL volatility ,more extreme losses and gains

The boxplot confirms this because large Position Traders show wider dispersion & more extreme outliers exist on both profit and loss sides

This suggests large traders rely on fewer high-conviction trades with larger exposure, accepting higher volatility and lower consistency in exchange for larger upside potential.

Meanwhile small position traders appear more conservative, maintaining higher win rates with smaller but more stable returns.

#### Segment 2 — Frequent vs Infrequent Traders

```python
trade_counts = trades.groupby("account").size()

threshold = trade_counts.median()

active_traders = pd.DataFrame({
    "trade_count": trade_counts
})

active_traders["activity_group"] = np.where(
    active_traders["trade_count"] >= threshold,
    "Frequent",
    "Infrequent"
)
```

```python
trades = trades.merge(
    active_traders["activity_group"],
    left_on="account",
    right_index=True
)
activity_perf = trades.groupby("activity_group").agg({
    "closed_pnl": "mean",
    "is_profit": "mean",
    "notional": "mean"
})

print(activity_perf)
```

#### Insights

Frequent traders achieve slightly better win rates, trade with larger average notionals ,but generate lower average PnL per trade

This may indicate Overtrading behavior reduces trade efficiency.

Infrequent traders, despite lower win rates, generate much larger average profits, implying better trade selection,stronger conviction,higher reward-to-risk opportunities

This aligns with common market behavior where selective trading can outperform excessive activity during volatile market conditions.

#### Segment 3 — Consistent Winners

```python
consistency = trades.groupby("account").agg({
    "is_profit": "mean",
    "closed_pnl": "std"
})

consistency.columns = ["win_rate", "pnl_volatility"]

consistency["segment"] = np.where(
    consistency["win_rate"] >= 0.6,
    "Consistent Winners",
    "Inconsistent"
)
```

```python
sns.boxplot(
    data=trades.merge(
        consistency["segment"],
        left_on="account",
        right_index=True
    ),
    x="segment",
    y="closed_pnl"
)

plt.title("PnL Distribution by Trader Segment")
plt.show()
```

#### Insights

The boxplot chart shows:

Inconsistent Traders : extremely wide PnL distribution, very large positive and negative outliers , heavy volatility

Consistent Winners : tighter PnL distribution , fewer extreme losses , more stable outcomes


This indicates that consistent winners prioritize stability and controlled risk rather than chasing large speculative profits.

Meanwhile Inconsistent traders experience higher variance, suggesting emotional or aggressive trading behavior.

The large negative outliers among inconsistent traders suggest poor risk management , inability to control downside during adverse market conditions

Overall, the findings suggest that profitability in volatile crypto markets is driven less by trade frequency and more by position management, trade selectivity, and effective downside control.

# Part C — “Actionable output”

## Strategy 1 — Reduce Position Size During Fear Periods

### During Fear or Extreme Fear market conditions, traders should reduce position sizes and avoid aggressive exposure, especially high-volatility traders.



The analysis showed that large position traders experiencing larger drawdowns ,lower win rates ,significantly higher PnL volatility.

Inconsistent traders had extreme negative outliers. Fear periods are typically associated with unstable market behavior

This suggests that large exposures become significantly riskier during fearful market conditions.

### Trading suggestions:

* Reduce position size by 30–50% during Fear periods
* Tighten stop losses
* Avoid averaging into losing positions
* Prioritize capital preservation over aggressive profit seeking

This strategy may reduce extreme downside events , improve long-term survivability & stabilize trader performance during volatile sentiment regimes

## Strategy 2 — Favor Selective High-Quality Trades Over Overtrading

### During highly emotional market periods, prioritize fewer high-conviction trades instead of increasing trade frequency.



The analysis showed frequent traders had higher activity , slightly higher win rates but lower average profitability

Infrequent traders generated substantially larger average PnL per trade

This indicates that trade quality matters more than trade quantity.

Excessive trading likely introduces:

* emotional decision-making
* transaction cost accumulation
* lower-quality setups

### Trading suggestions:

* Limit maximum daily trades
* Only enter trades meeting strict setup criteria
* Avoid revenge trading after losses
* Trade only during strong trend confirmations

This strategy may improve average trade quality ,reduce unnecessary market exposure ,increase risk-adjusted returns

## Strategy 3 — Use Sentiment as a Risk Regime Filter

### Increase directional exposure during Greed periods and prioritize defensive or short-biased positioning during Fear periods.



The behavioral analysis showed:

* sentiment influences trader positioning
* long/short ratios shift across Fear vs Greed periods
* market psychology strongly affects momentum

This implies:

market sentiment can act as a macro risk indicator.

During Greed:
allow larger trend-following positions ,favor momentum trades

During Fear:reduce leverage, hedge positions, increase cash allocation,favor short-term mean reversion trades
